```python
import requests
from bs4 import BeautifulSoup
import time

# トップページURL
url = "https://www.musashino-u.ac.jp/"

# アクセス
res = requests.get(url)
res.encoding = res.apparent_encoding  # ← 文字化け対策
print("ステータスコード：", res.status_code)  # 200ならOK

# HTMLを解析
soup = BeautifulSoup(res.text, "html.parser")

# <title>タグを取得
title_tag = soup.find("title")
print(f"<title>タグ：{title_tag}")
if title_tag:
    print(f"titleの文字列：{title_tag.text}")

# <h2>タグを全部取ってみる
h2_all = soup.find_all("h2")
print(f"<h2>タグのリスト：{h2_all}")  # <h2>タグオブジェクトのリスト
print(f"型：{type(h2_all)}")  # list型になってるか確認

# 各h2タグの中身を順番に出す
print("===== <h2>タグの中身 =====")
for h in h2_all:
    print(h.text.strip())

# aタグも試しに取ってみる
a_all = soup.find_all("a")
print(f"aタグの数：{len(a_all)}")

# 最初の5個くらいだけ表示
for a in a_all[:5]:
    print("リンク：", a.get("href"))

# 1秒待ってから次（負荷軽減）
time.sleep(1)
```

```python
# ...existing code...
start = "https://www.musashino-u.ac.jp/"
done = []   # 訪問したURLを入れる
data = {}   # URLとタイトルを入れる辞書

def crawl(url):
    if url in done:
        return
    done.append(url)
    res = requests.get(url, headers=headers)
    res.encoding = res.apparent_encoding
    soup = BeautifulSoup(res.text, 'html.parser')
    title = soup.title.string if soup.title else 'No Title'
    data[url] = title

    # 各ページで見出しを取得して URL と一緒に表示
    for tag in soup.find_all(['h2','h3','h4','h5']):
        print(f"URL: {url}  |  {tag.name}: {tag.get_text(strip=True)}")

    for link in soup.find_all('a', href=True):
        href = link['href']
        if href.startswith(start) and href not in done:
            crawl(href)

# クロール開始
crawl(start)
print(f"クロール完了（訪問数: {len(done)}）")
# ...existing code...
```

```python
with open("musashino_urls.txt", "w", encoding="utf-8") as f:
    for url, title in data.items():
        f.write(f"{url} → {title}\n")
        import time

time.sleep(1)  # 1秒休む

```
